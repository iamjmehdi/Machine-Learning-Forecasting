- a. Lambda (ùúÜ) in Ridge/Lasso Regression
- b. Epsilon (ùúÄ) & Budget (C) in Support Vector Machines

### Your Answer Here
a) The hyperparameter Lambda(ùúÜ) is a fixed positive constant. 
   The hyperparameter Lambda(ùúÜ) in Ridge/Lasso Regression is considered good if it offers the least overfitting.
   
   In Ridge Regression, we add a penalty by a way of a tuning parameter Lambda(ùúÜ), which is chosen using cross-validation. The idea is to make the fit small by making the residual sum or squares small, plus adding a shrinkage pentalty.
   
   In Lasso Regression, the penalty is the value of the sum of the absolute value of the coefficients. Lasso shrinks the coefficient estimates towards zero and it ha the effect of setting the variables exactly equal to 0, when Lambda(ùúÜ) is large enough (unlike Ridge). The tuning parameter is also chosen by cross-validation. 

b) In a Support Vector Machine the number Epsilon (ùúÄ) is a tunable hyperparameter.
The value of Epsilon (ùúÄ) defines a margin of tolerance where no penalty is given to errors. 

Furthermore, the Budget/Cost (C), simply put, is the idea that we optimize both the fit of the line to the data and penalize the amount of samples in the margin simultaneously, where C defines the weight of how much samples inside the margin contribute to the overall error.

The predictive goal of a supervised machine learning task is to find a model, which best describes the underlying population of data. There are 2 major sources of errors in such a task:
