```python
jet= plt.get_cmap('jet')
colors = iter(jet(np.linspace(0,1,10)))
 
def correlation(zacs_features1,variables, n_rows, n_cols):
    fig = plt.figure(figsize=(8,6))
    #fig = plt.figure(figsize=(14,9))
    for i, var in enumerate(variables):
        ax = fig.add_subplot(n_rows,n_cols,i+1)
        asset = zacs_features1.loc[:,var]
        ax.scatter(zacs_features1["ZORI"], asset, c = next(colors))
        ax.set_xlabel("ZORI")
        ax.set_ylabel("{}".format(var))
        ax.set_title(var +" ZORI")
    fig.tight_layout() 
    plt.show()
        
# Take the name of the last 6 columns of the zacs_features which are the model features
variables = zacs_features1.columns[-8:]
 
correlation(zacs_features1,variables,3,3)
```
